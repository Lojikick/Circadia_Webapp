```python
import mne
import yasa
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#great resources
```

```python
df_eeg = pd.read_csv('chris_eeg.csv')
```

```python
df_eeg =df_eeg.fillna(0)
```

```python
df_eeg[0:50]
```

```python
sample_data = df_eeg

sampling_rate = 256

ch_names = ['TP9', 'AF7', 'AF8', 'TP10']

time_series = []
time_series.append(np.array(sample_data['RAW_TP9']))
time_series.append(np.array(sample_data['RAW_AF7']))
time_series.append(np.array(sample_data['RAW_AF8']))
time_series.append(np.array(sample_data['RAW_TP10']))
time_series = np.array(time_series)
# channels = ['RAW_TP9', 'RAW_AF7', 'RAW_AF8', 'RAW_TP10']
info = mne.create_info(ch_names,sampling_rate)
raw = mne.io.RawArray(time_series, info)
```

```python
time_series
```

```python
raw
```

```python
sls = yasa.SleepStaging(raw, eeg_name="AF7", metadata=dict(age=21, male=False))
y_pred = sls.predict()
df_pred = pd.DataFrame({'Stage': y_pred})
df_pred.to_csv("my_hypno.csv")
```
