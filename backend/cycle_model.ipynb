```python
import mne
import yasa
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from yasa import Hypnogram
#great resources
```

```python
sample_data = pd.read_csv('sleep_eeg.csv')
sample_data =sample_data.fillna(0)
sampling_rate = 256
ch_names = ['TP9', 'AF7', 'AF8', 'TP10']
time_series = []
time_series.append(np.array(sample_data['ch1']))
time_series.append(np.array(sample_data['ch2']))
time_series.append(np.array(sample_data['ch3']))
time_series.append(np.array(sample_data['ch4']))
time_series = np.array(time_series)
info = mne.create_info(ch_names,sampling_rate)
raw = mne.io.RawArray(time_series, info)
sls = yasa.SleepStaging(raw, eeg_name="TP9", metadata=dict(age=21, male=False))
y_pred = sls.predict()
df_pred = pd.DataFrame({'Stage': y_pred})
hypnogram_predictions = np.array(df_pred["Stage"])
```

```python
hyp = Hypnogram(hypnogram_predictions, freq="24min")
```

```python
ax = hyp.plot_hypnogram(lw=2, fill_color="thistle")
```

```python
fig = yasa.plot_spectrogram(np.array(time_series[0]), 256, hypno=hypnogram_predictions)
```
